```python
import pandas as pd
```

```python
data = pd.read_excel('Amazon.xlsx')
data.head(1)
```

# 1.Basic Feature Extraction

```python
amazon = data[['reviews.text']]
amazon.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
amazon['Number_of_words']= amazon['reviews.text'].apply(lambda x:len(x.split()))
amazon.head(1)
```

```python
amazon['Number_of_characters'] = amazon['reviews.text'].apply(lambda y:len(y))
amazon.head(1)
```

```python
total_characters = amazon['Number_of_characters'].sum()
total_characters
```

```python
amazon['Average_word_length'] = amazon['Number_of_characters'].apply(lambda z:z/total_characters)
amazon.head(1)
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
stop
```

```python
amazon['Number_of_stop_words']=amazon['reviews.text'].apply(lambda word:len([word for word in word.split() if word not in stop]))
amazon.head(1)
```

```python
import string
punctuations = tuple(string.punctuation)
punctuations
```

```python
count = 0
a = 'aa@#bc'
for i in a:
    if i in string.punctuation:
        count = count+1
print(count)
```

```python
amazon['No_of_Punctuations'] = amazon['reviews.text'].apply(lambda x: len([x for x in x.split() if x in string.punctuation]))
amazon.head(1)
```

```python
amazon['Number_of_numerics'] = amazon['reviews.text'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
amazon.head(1)
```

```python
amazon['No_of_uppercase_words'] = amazon['reviews.text'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
amazon.head(1)
```

# 2.Basic Pre - Processing

```python
# converting to lower case and removing punctuations
amazon['clean_text'] = amazon['reviews.text'].str.lower().str.replace('[^a-z]',' ')
amazon.head(1)
```

```python
# remoing stop words
amazon['split_words'] = amazon['reviews.text'].apply(lambda x:' '.join(x for x in x.split() if x not in stop))
amazon.head(1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

# one gram 

```python
count_vect = CountVectorizer(max_df = 0.8,min_df = 0.2,stop_words = 'english')
train_vect = count_vect.fit_transform(amazon['split_words'])
train_vect
```

```python
count_one_gram = pd.DataFrame(train_vect.toarray(),columns = count_vect.get_feature_names())
count_one_gram.head()
```

```python
word_count = pd.DataFrame(count_one_gram.sum())
word_count = word_count.rename(columns = {0:'Word_Count'})
word_count = word_count.sort_values(by = 'Word_Count',ascending = False)
word_count.head()
```

# two gram

```python
two_gram = CountVectorizer(max_df = 0.9,min_df = 0.02,ngram_range = (2,2), stop_words = 'english')
two_gram_train_vect = two_gram.fit_transform(amazon['split_words'])
two_gram_train_vect
```

```python
DTM_two_gram = pd.DataFrame(two_gram_train_vect.toarray(),columns = two_gram.get_feature_names())
DTM_two_gram.head()
```

```python
count_two_gram = pd.DataFrame(DTM_two_gram.sum())
count_two_gram = count_two_gram.rename(columns = {0:'Word_Count'})
count_two_gram = count_two_gram.sort_values(by = 'Word_Count',ascending = False)
count_two_gram.head()
```

# three gram

```python
three_gram = CountVectorizer(max_df = 0.8, min_df = 0.02,ngram_range = (3,3),stop_words = 'english')
three_gram_vect = three_gram.fit_transform(amazon['split_words'])
three_gram_vect
```

```python
DTM_three_gram = pd.DataFrame(three_gram_vect.toarray(),columns = three_gram.get_feature_names())
DTM_three_gram.head()
```

```python
count_three_gram = pd.DataFrame(DTM_three_gram.sum())
count_three_gram = count_three_gram.rename(columns = {0:'Word_Count'})
count_three_gram = count_three_gram.sort_values(by = 'Word_Count', ascending = False)
count_three_gram.head()
```

```python
count_three_gram['Word_Count'].nlargest(3)
```

# Tf - Idf Bag of words

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer( ngram_range = (3,3), stop_words = 'english')
tfidf_vect = tfidf.fit_transform(amazon['split_words'])
tfidf_vect
```

```python
DTM_tfidf_three_gram = pd.DataFrame(tfidf_vect.toarray(),columns = tfidf.get_feature_names())
DTM_tfidf_three_gram.head()
```

```python
score_tfidf = pd.DataFrame(DTM_tfidf_three_gram.sum())
score_tfidf = score_tfidf.rename(columns = {0:'Score'})
score_tfidf = score_tfidf.sort_values(by = 'Score', ascending = False)
score_tfidf.head()
```
